```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.rand(100)  # x-axis values
y = np.random.rand(100)  # y-axis values
colors = np.random.rand(100)  # colors for each point
sizes = 100 * np.random.rand(100)  # sizes for each point

# Plotting the scatter plot with customizations
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot Example')

# Adding colorbar
cbar = plt.colorbar()
cbar.set_label('Color')

# Displaying the plot
plt.show()
```
